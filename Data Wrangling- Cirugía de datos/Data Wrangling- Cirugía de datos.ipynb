```python
#Funcion para generar una lista de n numeros enteros dentro de un intervalo
def randint_list (n, a, b):
    x=[]
    for i in range (n):
        x.append(np.random.randint(a,b))
    return x 
```

# Simulacion de Monte Carlo

```python
def valor_pi (n, nexp):
    pi_avg= 0
    pi_value_list= []
    for i in range(nexp):
        value=0
        x= np.random.uniform(0,1,n).tolist()
        y= np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z= np.sqrt(x[j]*x[j]+y[j]*y[j])
            if z <=1:
                value +=1
        float_value= float(value)
        pi_value= (float_value*4)/n
        pi_value_list.append(pi_value)
        pi_avg += pi_value
    pi =pi_avg/nexp
    print(pi)
    fig = plt.plot(pi_value_list)
    return(pi, fig)
        
```

```python
valor_pi(1000, 100)
```
