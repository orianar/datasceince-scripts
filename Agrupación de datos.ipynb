# Agregacion de datos por categoria

```python
%config IPCompleter.greedy=True
import numpy as np
import pandas as pd
```

```python
genero= ["female","male"]
```

```python
nivel_ingresos = ["Bajo", "Medio", "alto"]
```

```python
n=500
genero_data=[]
ingresos_data=[]
```

```python
for i in range (0, 500):
    genero_data.append(np.random.choice(genero))
    ingresos_data.append(np.random.choice(nivel_ingresos))
```

```python
altura= 160 + 30 * np.random.randn(n)
edad =30 + 12*np.random.randn(n)
ingresos=14000 + 2000 *np.random.randn(n)
peso=65 + 25 *np.random.randn(n)
```

```python
data= pd.DataFrame(
    {'Genero':genero_data,
     'Ingresos':ingresos_data,
     'Estatura':altura,
     'Ganancia': ingresos,
     'peso':peso,
    }
)
```

```python
data.tail()
```

# Agrupacion de datos

```python
doble_grupo.aggregate(
    {
        'Estatura':np.sum,
        'Ganancia':np.mean,
        'peso':np.std,
    }
)
```

```python
doble_grupo.aggregate(
    {
        'Estatura': np.mean,
        'Ganancia': lambda h : np.mean(h)/np.std(h)
    }
)
```
