## Clustering jerarquico

```python
import matplotlib.pyplot as plt
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df[movies].head()
```

```python
j= linkage(df[movies], "ward")
j
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(j, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```

```python
j = linkage(df[movies], "complete")
j
plt.figure(figsize=(25,10))
plt.title("Dendrograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(j, leaf_rotation=90., leaf_font_size=10.0)
plt.show()
```
