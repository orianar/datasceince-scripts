```python
import numpy as np
```

```python
out_athletes= np.random.choice(data_main["Athlete"], size= 6, replace= False)
```

```python
filter_Athletes=(~data_country_dp["Athlete"].isin (out_athletes)) & (data_country_dp["Athlete"]!= "Michael Phelps")
data_contry_dlt= data_country_dp[filter_Athletes] 
data_sport_dlt= data_sport_dp[(~data_sport_dp["Athlete"].isin (out_athletes)) & 
                              (data_sport_dp["Athlete"]!= "Michael Phelps")]
```

```python
data_sport_dlt
```

```python
## Inner 
meger_inner= pd.merge (left= data_main, right=data_contry_dlt,
                      how="inner", left_on= "Athlete", right_on="Athlete")
```

```python
len(meger_inner)
```

```python
## Left
meger_Left= pd.merge (left= data_main, right=data_contry_dlt,
                      how="left", left_on= "Athlete", right_on="Athlete")
```

```python
len(meger_Left)
```

```python
## right
meger_right= pd.merge (left= data_main, right=data_contry_dlt,
                      how="right", left_on= "Athlete", right_on="Athlete")
```
